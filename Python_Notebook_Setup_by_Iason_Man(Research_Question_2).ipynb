Codebook 1

image.png

```python
#Manage data for variable 1
df['H4PE33'].value_counts()
```

```python
df['Avoid_Problems'] = df['H4PE33']
df['Avoid_Problems'] = df['Avoid_Problems'].loc[ df['Avoid_Problems'] !=' ']
df['Avoid_Problems'] = df['Avoid_Problems'].loc[ df['Avoid_Problems'] !='8']
df['Avoid_Problems'] = df['Avoid_Problems'].loc[ df['Avoid_Problems'] !='6']
df['Avoid_Problems'].value_counts()
```

```python
df['Avoid_Problems'].replace(['1','2','3','4','5'] , ['Stongly Agree','Agree','Neither Agree nor Disagree','Disagree','Strongly Disagree'], inplace = True)
df['Avoid_Problems'].value_counts()
```

```python
#Graphs and tables or statistics
df['Avoid_Problems'].describe()
```

```python
df['Avoid_Problems'].value_counts().plot(kind ='pie', autopct='%1.1f%%')
```

```python
df['Avoid_Problems'].value_counts(normalize = True).plot(kind ='bar')
```

```python
#Manage data for variable 2
df['H4RE11'].value_counts()
```

```python
df['belief'] = df['H4RE11']
df['belief'] = df['belief'].loc[ df['belief'] !=' ']
df['belief'] = df['belief'].loc[ df['belief'] !='8']
df['belief'] = df['belief'].loc[ df['belief'] !='6']
df['belief'].value_counts()
```
